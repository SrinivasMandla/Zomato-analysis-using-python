**CONVERT THE DATATYPE OF COLUMN-RATE**

````python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value);
````

````python
dataframe['rate']=dataframe['rate'].apply(handleRate)
print (dataframe['rate'])
````

**Q5. WHICH TYPE OF RESTAURENT RECEVIED
MORE OFFLINE ORDERS, SO THAT ZOMATO CAN PROVIDE CUSTOMERS WITH SOME GOOD OFFERS:**

````python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu')
plt.title("heatmap")
plt.xlabel("online_order")
plt.ylabel("listed_in(type)")
plt.show()
````

**5)ANSWER: DINNING RESTAURENTS PRIMARILY ACCEPT OFFLINE ORDERS, WHEREAS CAFES PRIMARILY RECEIVE ONLINE ORDERS. THIS SUGGESTS THAT CLIENTS PREFER ORDERS IN PERSON AT RESTAURENTS, BUT PREFER ONLINE ORDERING AT CAFES.**
